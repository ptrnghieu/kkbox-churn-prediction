```python
import requests
import pandas as pd

resp = requests.get(
    "https://api.delphi.cmu.edu/epidata/fluview/",
    params={
        "regions": "nat",      # start with national, then expand to state
        "epiweeks": "202001-202452"
    }
)
data = resp.json()

if data["result"] == 1:
    df = pd.DataFrame(data["epidata"])
    print(df.shape)
    print(df.head())
else:
    print("Error:", data["message"])
```

```python
import requests
import pandas as pd
import time

# State region codes used by Delphi
STATES = [
    "al","ak","az","ar","ca","co","ct","de","fl","ga",
    "hi","id","il","in","ia","ks","ky","la","me","md",
    "ma","mi","mn","ms","mo","mt","ne","nv","nh","nj",
    "nm","ny","nc","nd","oh","ok","or","pa","ri","sc",
    "sd","tn","tx","ut","vt","va","wa","wv","wi","wy"
]

# Always inspect the raw response first before building a DataFrame
resp = requests.get(
    "https://api.delphi.cmu.edu/epidata/fluview/",
    params={
        "regions": ",".join(STATES),   # comma-separated, not just "state"
        "epiweeks": "202001-202452"
    }
)
data = resp.json()

# Check what the API actually returned
print("result code:", data["result"])
print("message:", data["message"])
print("num records:", len(data.get("epidata", [])))
```

```python
df = pd.DataFrame(data["epidata"])
print(df.shape)
print(df.isnull().sum())
```
